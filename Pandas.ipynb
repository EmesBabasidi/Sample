# Introduction to Pandas

```python
#To use pandas just like numpy, you import pandas as shown below
import pandas as pd
import numpy as np
```

```python
#data series contain only one row or column
#data frame is a combination of with multiple rows and columns
```

```python
#creating a data frame
#let's first create some arrays from the lists below:
muhsin = ["Muhsin",27,"Graduate"]
mutawakkil = ["Mutawakkil",23,"Penultimate level"]
mahbub = ["Mahbub", 20, "Fresher"]
adnan =["Adnan", 17, "Awaiting Admission"]
myArr = np.array([muhsin,mutawakkil,mahbub,adnan])
print(myArr)
```

```python
#we now use pd.DataFrame to create the data frame
#this function recieves the data parameter(ARRAY), index(Row titles/headings), and columns(column titles/headings)
#the number of elements in the index must be equal to the number of rows in the given array
#also the number of elements in the columns must be equal to the number of columns in the given array
dataFrame = pd.DataFrame(data = myArr, index = [1,2,3,4], columns = ["Names","Age","Educational Status"])

#other parameters that the function recieves will be discussed later
```

```python
#the function .head returns 'n' number of rows as passed to it, if no value is passed it returns all rows available
dataFrame.head()
```

```python
#Example of when a number is passed to .head function
dataFrame.head(1)
```

```python
#We can convert a dataframe we've created here into a csv file using .to_csv(). Example
dataFrame.to_csv('Example1.csv')
#and once executed we can find our csv file in the same folder with our program
```

# Accessing elements from a dataframe 

```python
#using .loc[] which access only row elements, and can recieve only one parameter at a time. Example:
dataFrame.loc[2]
```

```python
#using .iloc[] which is similar to accessing multidimensional arrays in numpy
#[row index,column index] in both cases a:b mean from index "a" to index "b-1"
#When no number is passed for example [:] it means accesss everything
#when no ending is specified for example [a:] it means from index "a" to the last element 
#Example
dataFrame.iloc[2:,0:2]
```

```python
#by calling name of column(s), example:
#for more than one column, use two square brackets instead of one
dataFrame[['Names','Age','Educational Status']]
```

```python
# converting a dataframe into an array using .values
dataFrame.values
```

```python
# you can select which part of the dataframe to convert using .iloc[:,:].values
dataFrame.iloc[2:,0:2].values
```

```python
#using .nul() to check if there is a missing data from the dataframe especially if imported, example:
dataFrame.isnull()
```

```python
dataFrame.isnull().sum()
#this wiil return the number of omitted data from each column
```

# Reading csv files and
# Using some inbuilt functions to perform operations

```python
#to import, we use pd.read_csv() and pass the file name in a qoutes
# a csv must not be comma seperated only, though the means of seperation must be indicated. 
#Example: if it's a semi colon seperated,it will be pd.read_csv('fifa24.csv', sep = ';')
df = pd.read_csv('fifa24.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
#as it can be seen there are 248 missing data from our imported csv file from the Contract_end column
#the .sum() is used so that we can see the number of missing elements, else we will just be seeing TRUE/FALSE 
#Try removing .sum() and see the effect
```

```python
#.info gives you information of a csv file
df.info()
```

```python
#.decribe function generate descriptive statistics.
df.describe()
```

```python
#The function .value_counts() return the number of occurrance of an element in a given data set
#a column or row can be specified as shown below.The function recieves other parameters 
df['Positions'].value_counts(sort = True, ascending =False)
```

```python
#a condional search can be made within a dataset as shown below
df[df['National_team'] == 'Nigeria']
```

```python
# A logical search can also be made & = and, | = or, ~ = not. Example
df[(df['Age'] <= 20) & (df['Overall'] > 80) & (df['Potential_overall'] >= 90)]
```

```python
#the "isin()" function
df[df["National_team"].isin(['Nigeria','England','Poland'])]
```

```python
df['Age'].count()
```

# Dealing with JSON files 

```python
#recall, df =  pd.read_csv('fifa24.csv'). we can convert this dataframe df to json format or any other format using
#.to_format('Name_of_file.format')
#orient is like formatting the json files, the types of format are split,records,index,columns,values,table
df.to_json('fifa24.json',orient = 'records')
```

```python
#to import, we use pd.read_json() and pass the file name in a qoutes
df1 =  pd.read_json('fifa24.json')
```

```python
df1.head(10)
```

```python
# .sample is used to obtain a random sample from the dataframe
# frac used here will return 10% random sample from the dataframe, frac cannot be used with "n" and vice versa.
# if frac is > 1, replace should be = true, as by default it is false 
# replace allows the sampling(i.e repeating) a row more than once.    
#YOU CAN CHECK OTHER ARGUEMENTS THAT CAN BE PASSED TO THE sample() function
df1.sample(frac = .1)
```
